The second stage of the project focuses on preparing the dataset for analysis by improving its quality and consistency. Raw datasets often contain missing values, incorrect data types, duplicate records, or variables that require additional transformation before meaningful insights can be extracted.

During this stage, the dataset is cleaned by handling missing values, verifying duplicate records, correcting data types where necessary, and creating new derived features that enhance the analytical process. #### 1) HANDLING MISSING VALUES

```python
# displaying the number of missing values in each column.
df.isnull().sum()
```

```python
# replace missing values in the O3, CO, SO2,NO2 column with its median value.
# median is used because it is data types are float datatype.

df["O3"] = df["O3"].fillna(df["O3"].median())

# replace missing values in the CO column with its median value.
df["CO"] = df["CO"].fillna(df["CO"].median())

# replace missing values in the SO2 column with its median value.
df["SO2"] = df["SO2"].fillna(df["SO2"].median())

# replace missing values in the NO2 column with its median value.
df["NO2"] = df["NO2"].fillna(df["NO2"].median())
```

```python
# verifying that all missing values have been handled
df.isnull().sum()
```

#### 2) REMOVING DUPLICATE RECORDS

```python
# counting the number of duplicate rows in the dataset.
duplicate_rows = df.duplicated().sum()
print("Duplicate Rows:", duplicate_rows)
```

```python
#Removing duplicate rows from the dataset.
df = df.drop_duplicates()
```

```python
# verifying no duplicate rows remain.
print("Duplicate Rows:", df.duplicated().sum())
```

#### 3) Correcting Data Types

```python
df.dtypes
```

```python
# convert the Hour column from object to integer
# hour column is stored as a string in the format "HH:MM" so extracting only the hour (HH) and convert it to an integer.
df["Hour"] = df["Hour"].str.split(":").str[0].astype(int)
```

```python
# displaying the updated data types.
print(df.dtypes)
```

```python
df
```

```python
# Remove the alphabetical prefixes such as a_, b_, c_, d_, e_, and f from the AQI_Class values.

df["AQI_Class"] = df["AQI_Class"].str.replace(
    r"^[a-z]_",
    "",
    regex=True
)

# Replace underscores with spaces to make the category names more readable.

df["AQI_Class"] = df["AQI_Class"].str.replace("_", " ")

# Display the cleaned AQI categories.

print(df["AQI_Class"].unique())
```

#### 4) Creating Derived Columns

```python
# combining the year, month and day columns into a single date column.
df["Date"] = pd.to_datetime(df[["Year", "Month", "Day"]])
```

```python
# extract the month name from the Date column.
df["Month_Name"] = df["Date"].dt.month_name()
```

```python
# extracting the day of the week from the date column.
df["Day_Name"] = df["Date"].dt.day_name()
```

```python
# identifying whether the record falls on a weekend.
df["Weekend"] = df["Day_Name"].isin(["Saturday", "Sunday"])
```

```python
# categorize each record into a season based on the month.
def season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Summer"
    elif month in [6, 7, 8, 9]:
        return "Monsoon"
    else:
        return "Post-Monsoon"

df["Season"] = df["Month"].apply(season)
```

#### 5) Filtering Data

```python
# filter records where AQI is greater than 200.
high_aqi = df[df["AQI"] > 200]

# display the first five high pollution records.
high_aqi.head()
```

```python
df = df.drop(columns=["Filename"]) #removing unnecessary columns
```

```python
# display all unique monitoring locations.
print(df["Location"].unique())
```

```python
df = df[df["Location"] != "Biratnagar, Nepal"]
```

```python
# Display the updated number of rows and columns after removing Nepal records.

print("Dataset Shape:", df.shape)
```

```python
# categorize AQI values into pollution levels.

df["Pollution_Level"] = pd.cut(
    df["AQI"],
    bins=[0, 100, 200, 300, 500],
    labels=["Good", "Moderate", "Poor", "Severe"]
)
```

## Key Insights

### Insight 1: Variation in Air Quality Across Locations

The location-wise analysis using average AQI values revealed noticeable differences in air quality across the selected Indian locations. **ITO, Delhi** recorded the highest average AQI (**324.89**), indicating comparatively poorer air quality, while **Bengaluru** recorded the lowest average AQI (**66.14**). New Ind Town, Faridabad (230.00) and Knowledge Park, Greater Noida (226.35) also showed elevated pollution levels, suggesting that the Delhi-NCR region is a clear pollution hotspot. These variations may be influenced by differences in traffic, population density, industrial activities, and other local environmental factors.

---

### Insight 2: Relationship Between Pollutants and AQI

The correlation analysis was used to examine the relationship between AQI and major air pollutants. Among the pollutants analyzed, **PM2.5** showed the strongest positive correlation with AQI, with a correlation value of **0.85**, followed by **PM10 (0.72)** and **NO2 (0.55)**. This indicates that higher concentrations of particulate matter are generally associated with higher AQI values and poorer air quality. However, correlation indicates an association and does not necessarily prove a direct cause-and-effect relationship.

---

### Insight 3: Distribution of AQI Categories

The frequency analysis of AQI categories showed that **Very Unhealthy** was the most frequently recorded air quality category in the dataset, closely followed by **Unhealthy for Sensitive Groups**. This indicates that a considerable proportion of the observations experienced hazardous air quality conditions. In comparison, **Moderate** occurred least frequently during the observed period, showing that clean-to-moderate air days were relatively uncommon in the monitored locations.

---

### Insight 4: Seasonal Variation in Air Quality

The seasonal analysis showed variations in average AQI and pollutant concentrations across different seasons. **Post-Monsoon** recorded the highest average AQI (**230.00**), followed by **Winter (195.53)**, while **Summer** recorded the lowest (**117.42**). This suggests that seasonal conditions may influence the accumulation and dispersion of air pollutants. Factors such as rainfall, wind conditions, temperature, and changes in human activities may contribute to these variations. It should be noted that the Post-Monsoon records come from a single location (Faridabad), so the Winter figure is more representative of the overall dataset.

---

### Insight 5: Variation in Pollutant Concentrations

The descriptive statistical analysis showed that **PM2.5** had the highest variability among the pollutants analyzed, with a standard deviation of **136.74** and values ranging from **4 to 500**. This indicates that its concentration changed considerably across different locations and observation periods. **CO** also showed high variability (standard deviation of 114.67). The presence of unusually high values may represent periods of increased emissions or temporary pollution events.

---

### Insight 6: Monthly Air Quality Trends

The month-wise analysis identified changes in average AQI across the observation period. **October** recorded the highest average AQI (**230.00**), whereas **March** recorded the lowest (**117.42**), with February in between (**195.53**). This indicates that air quality conditions were not consistent throughout the year and may be influenced by seasonal and environmental changes such as post-monsoon crop burning and winter atmospheric inversion, which are known to worsen air quality in North India during October–February.

---

### Overall Finding

The analysis demonstrates that air quality varies significantly across locations, seasons, and time periods. Delhi-NCR locations consistently recorded the poorest air quality, while southern cities such as Bengaluru recorded the best. Particulate matter (PM2.5 and PM10) showed the strongest relationship with AQI, making it the most important pollutant to monitor and control. These findings highlight the importance of location-specific air quality monitoring and targeted measures to reduce the pollutants most strongly associated with poor air quality.
